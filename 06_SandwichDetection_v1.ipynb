# Heuristics 1,4,6a

Add count (size) of sub-frames grouped by same values in same block in same LP. 'Count_1'-value >1 means that there are 1+ observations with the same value in same block and LP, which is a great indicator for an attack.

```python
df1['Count_1']=df1.groupby(['block_number','poolId', 'value'],as_index=True)['value'].transform('size').astype('float') 
```

Adding ncount of unique transactions-hashes per sub-frame block, LP, value. Only if ncount > 1 we have 2 seperate transactions with these criteria, because a single transaction can have more then 2 observations.
* https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.GroupBy.count.html
* https://www.tutorialspoint.com/count-unique-values-per-groups-in-python-pandas


```python
### Archived
#df1['unique_transactions'] = df1.groupby(['block_number','poolId', 'value'],as_index=True)['transaction_hash'].transform('count').astype('float')


```

Filtering for 'Count_1' > 1 in grouped sub-frames by transaction_hashes per block and LP. This results in a df that contains all observations (both observations of a transaction because mean is > 1 if only 1 of them is larger 1) that have same values per sub-frame by block, LP.


```python
# Filtering for count_1 mean > 1 per sub-frame
# x in lambda function refers to to indiviudal sub-frames
df_filtered = df1.groupby(['block_number','poolId', 'transaction_hash'],as_index=True).filter(lambda x: x['Count_1'].mean() > 1)


```

# Heuristics 3 & 7

By counting the unique hashes per sub-frame by block, LP, value we can ensure that each AT1 and AT2 have different hashes. The count needs to be >= 2.

* Assumption: this heuristic will not remove many observations: 188'663 observations removed => double check
* ATTENTION: We narrow down to attacks with exctly 2 transactions, no splitup attacks. This represents heuristic 7.
* More than 4 observations are allowed but only applicable in rare cases. Check with df_filtered2 = df_filtered2.groupby(['transaction_hash'],as_index=True).filter(lambda x: len(x)> 4)

```python
# Add column with unique count of transaction_hash per sub-frame
df_filtered['Count_nHash']=df_filtered.groupby(['block_number','poolId', 'value'],as_index=True)['transaction_hash'].transform('nunique').astype('float')
```

```python
# filtering 'Count_nHash'-mean == 1.5 (formerly >1) per sub-frame by 'transaction_hash' means that there are exact 2 (formerly at least 2) different transaction_hashes, which is heuristic nr. 3
df_filtered2 = df_filtered.groupby(['transaction_hash'],as_index=True).filter(lambda x: x['Count_nHash'].mean()  == 1.5)

```

# Heuristic 5

### Creating 'within_transaction_order'
Creating additional column 'within_transaction_order' indicating the position of the observation within the transaction based on log_index.
Following procedure:

* sort by 'log_index'
* groupby ('block_number','poolId'), 'transaction_hash' (groupby preserves the order of rows within each group: https://sparkbyexamples.com/pandas/pandas-groupby-sort-within-groups/)
* add an enumerator starting at 1 (using add(1))

ATTENTION: Similar result could be achieved by simply using 'log_index'. However, new column is crated for better transparency and having acount of observations per group

```python
df_filtered2['within_transaction_order'] = df_filtered2.sort_values(['log_index'],ascending=True).groupby(['transaction_hash'],as_index=True).cumcount().add(1)
```
